```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset_name = "Bitcoin"
experiments_dir = f"/home/vahid/Projects/pythonProject/DynamicFraudGNN/experiments/edge_level/{dataset_name}/2024-08-28_12:30:08"

metric_files = {file: pd.read_csv(f"{experiments_dir}/{file}/lightning_logs/metrics.csv") for file in os.listdir(experiments_dir)}
metric_files.keys()
```

```python
def plot_metrics(metrics):
    aggreg_metrics = []
    agg_col = "epoch"
    for i, dfg in metrics.groupby(agg_col):
        agg = dict(dfg.mean())
        agg[agg_col] = i
        aggreg_metrics.append(agg)

    df_metrics = pd.DataFrame(aggreg_metrics)
    df_metrics[["train_loss", "val_loss"]].plot(
        grid=True, legend=True, xlabel="Epoch", ylabel="Loss"
    )
    df_metrics[["train_avg_pr", "val_avg_pr"]].plot(
        grid=True, legend=True, xlabel="Epoch", ylabel="avg_pr"
    )

    plt.show()
```

```python
plot_metrics(metric_files["timestamp_3"])
```

```python
version = 28
metrics = pd.read_csv(f"../saved_models/GNNs/edge_level/lightning_logs/version_{version}/metrics.csv")
plot_metrics(metrics)
```
